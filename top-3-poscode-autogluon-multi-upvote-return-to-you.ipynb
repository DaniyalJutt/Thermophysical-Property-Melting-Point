```python
import time
import shutil
import warnings
import numpy as np
import pandas as pd
from tqdm import tqdm
import multiprocessing
from rdkit import Chem
from rdkit.Chem import Descriptors
from autogluon.tabular import TabularPredictor
from rdkit.Chem.rdMolDescriptors import GetMorganFingerprintAsBitVect, GetMACCSKeysFingerprint

# Ignore all warnings to keep the output clean
warnings.filterwarnings("ignore")

# Start timing the script
start_time = time.time()

print("Loading data...")
try:
    train_df = pd.read_csv("/kaggle/input/melting-point/train.csv")
    test_df = pd.read_csv("/kaggle/input/melting-point/test.csv")
    sample_submission = pd.read_csv("/kaggle/input/melting-point/sample_submission.csv")
    print("Data loaded successfully.")
except FileNotFoundError as e:
    print(f"Error: {e}. Make sure the data is in the correct directory.")
    exit()
```

<!-- ================= BADGES: Important RDKit Functions ================= -->
<div style="display:flex; gap:10px; flex-wrap:wrap; justify-content:center; margin:12px 0; font-family:Inter, system-ui, 'Segoe UI', Roboto, Arial, sans-serif;">
  <span style="background:#e2e8f0; color:#0f172a; padding:6px 14px; border-radius:999px; font-size:13px; border:1px solid #cbd5e1;">Chem.MolFromSmiles</span>
  <span style="background:#dbeafe; color:#1e3a8a; padding:6px 14px; border-radius:999px; font-size:13px; border:1px solid #bfdbfe;">Descriptors</span>
  <span style="background:#dcfce7; color:#14532d; padding:6px 14px; border-radius:999px; font-size:13px; border:1px solid #bbf7d0;">GetMorganFingerprintAsBitVect</span>
  <span style="background:#fde68a; color:#92400e; padding:6px 14px; border-radius:999px; font-size:13px; border:1px solid #fcd34d;">GetMACCSKeysFingerprint</span>
</div>


```python

import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors, rdMolDescriptors, Crippen
from rdkit.Chem import AllChem
from rdkit.Chem.rdMolDescriptors import GetMorganFingerprintAsBitVect, GetMACCSKeysFingerprint
from tqdm import tqdm
import multiprocessing
import warnings
warnings.filterwarnings('ignore')

# Identify features and target
TARGET = "Tm"
ID_COL = "id"
SMILES_COL = "SMILES"

print(f"Train data shape: {train_df.shape}")
print(f"Test data shape: {test_df.shape}")

class EnhancedMolecularFeatures:
    def __init__(self):
        # No problematic imports - everything works
        pass
    
    def generate_3d_conformer(self, mol):
        """Generate 3D conformer for geometric descriptors"""
        try:
            mol = Chem.AddHs(mol)
            AllChem.EmbedMolecule(mol, randomSeed=42)
            AllChem.MMFFOptimizeMolecule(mol)
            return mol
        except:
            return None
    
    def get_3d_descriptors(self, mol_3d):
        """Calculate 3D geometric descriptors"""
        features = {}
        if mol_3d is None:
            return {f'3D_{key}': 0 for key in [
                'PMI1', 'PMI2', 'PMI3', 'NPR1', 'NPR2', 'RadiusOfGyration',
                'InertialShapeFactor', 'Eccentricity', 'Asphericity', 'SpherocityIndex'
            ]}
        
        try:
            features['3D_PMI1'] = rdMolDescriptors.PMI1(mol_3d)
            features['3D_PMI2'] = rdMolDescriptors.PMI2(mol_3d)
            features['3D_PMI3'] = rdMolDescriptors.PMI3(mol_3d)
            features['3D_NPR1'] = rdMolDescriptors.NPR1(mol_3d)
            features['3D_NPR2'] = rdMolDescriptors.NPR2(mol_3d)
            features['3D_RadiusOfGyration'] = rdMolDescriptors.RadiusOfGyration(mol_3d)
            features['3D_InertialShapeFactor'] = rdMolDescriptors.InertialShapeFactor(mol_3d)
            features['3D_Eccentricity'] = rdMolDescriptors.Eccentricity(mol_3d)
            features['3D_Asphericity'] = rdMolDescriptors.Asphericity(mol_3d)
            features['3D_SpherocityIndex'] = rdMolDescriptors.SpherocityIndex(mol_3d)
        except:
            features = {f'3D_{key}': 0 for key in [
                'PMI1', 'PMI2', 'PMI3', 'NPR1', 'NPR2', 'RadiusOfGyration',
                'InertialShapeFactor', 'Eccentricity', 'Asphericity', 'SpherocityIndex'
            ]}
        
        return features
    
    def get_enhanced_fingerprints(self, mol):
        """Generate enhanced fingerprints"""
        features = {}
        
        # Multiple Morgan fingerprint radii
        for radius in [2, 3, 4]:
            fp = GetMorganFingerprintAsBitVect(mol, radius=radius, nBits=512)
            for i in range(fp.GetNumBits()):
                features[f"Morgan_R{radius}_{i}"] = int(fp.GetBit(i))
        
        # MACCS Keys
        fp_maccs = GetMACCSKeysFingerprint(mol)
        for i in range(fp_maccs.GetNumBits()):
            features[f"MACCS_{i}"] = int(fp_maccs.GetBit(i))
        
        return features
    
    def get_functional_groups(self, mol):
        """Calculate functional group counts"""
        features = {}
        
        # Important patterns for melting point
        patterns = {
            'Hydroxyl': '[OH]',
            'Carbonyl': '[CX3]=[OX1]',
            'Carboxyl': '[CX3](=O)[OX2H1]',
            'Amine': '[NX3;!$(NC=O)]',
            'Ether': '[OD2]([#6])[#6]',
            'Ester': '[#6][CX3](=O)[OX2H0][#6]',
            'Amide': '[NX3][CX3](=[OX1])[#6]',
            'Benzene': 'c1ccccc1',
            'Halogen_F': '[F]',
            'Halogen_Cl': '[Cl]',
            'Halogen_Br': '[Br]'
        }
        
        for name, smarts in patterns.items():
            pattern = Chem.MolFromSmarts(smarts)
            if pattern:
                matches = mol.GetSubstructMatches(pattern)
                features[f'FG_{name}'] = len(matches)
            else:
                features[f'FG_{name}'] = 0
        
        return features
    
    def get_advanced_descriptors(self, mol):
        """Calculate advanced molecular descriptors"""
        features = {}
        
        try:
            features['CrippenLogP'] = Crippen.MolLogP(mol)
            features['CrippenMR'] = Crippen.MolMR(mol)
            features['NumAromaticRings'] = rdMolDescriptors.CalcNumAromaticRings(mol)
            features['NumAliphaticRings'] = rdMolDescriptors.CalcNumAliphaticRings(mol)
            features['FractionCsp3'] = rdMolDescriptors.CalcFractionCsp3(mol)
            features['BertzCT'] = rdMolDescriptors.BertzCT(mol)
            features['BalabanJ'] = rdMolDescriptors.BalabanJ(mol)
            features['Kappa1'] = rdMolDescriptors.Kappa1(mol)
            features['Kappa2'] = rdMolDescriptors.Kappa2(mol)
            features['Chi0'] = rdMolDescriptors.Chi0(mol)
            features['Chi1'] = rdMolDescriptors.Chi1(mol)
        except:
            features = {key: 0 for key in ['CrippenLogP', 'CrippenMR', 'NumAromaticRings', 
                       'NumAliphaticRings', 'FractionCsp3', 'BertzCT', 'BalabanJ',
                       'Kappa1', 'Kappa2', 'Chi0', 'Chi1']}
        
        return features

def generate_enhanced_features_from_smiles(smiles_list):
    """Generate enhanced features from SMILES list"""
    feature_gen = EnhancedMolecularFeatures()
    features = []
    
    for smiles in tqdm(smiles_list, desc="Processing SMILES"):
        mol = Chem.MolFromSmiles(smiles)
        feature_dict = {}
        
        if mol is not None:
            # Original RDKit descriptors - direct calculation (NO IMPORT ISSUES)
            try:
                for desc_name, desc_func in Descriptors.descList:
                    try:
                        feature_dict[desc_name] = desc_func(mol)
                    except:
                        feature_dict[desc_name] = 0
            except:
                pass
            
            # 3D descriptors
            mol_3d = feature_gen.generate_3d_conformer(mol)
            feature_dict.update(feature_gen.get_3d_descriptors(mol_3d))
            
            # Enhanced fingerprints
            feature_dict.update(feature_gen.get_enhanced_fingerprints(mol))
            
            # Functional groups
            feature_dict.update(feature_gen.get_functional_groups(mol))
            
            # Advanced descriptors
            feature_dict.update(feature_gen.get_advanced_descriptors(mol))
        
        features.append(feature_dict)
    
    return features

def parallel_enhanced_feature_generation(smiles_list, num_workers):
    """Generate enhanced features in parallel"""
    chunks = np.array_split(smiles_list, num_workers)
    with multiprocessing.Pool(processes=num_workers) as pool:
        results = pool.map(generate_enhanced_features_from_smiles, chunks)
    
    return [item for sublist in results for item in sublist]

# Get number of cores
num_cores = multiprocessing.cpu_count()
print(f"Using {num_cores} CPU cores for enhanced feature generation.")

# Generate enhanced features for both training and test sets
print("ğŸš€ Generating enhanced molecular features...")
train_features_list = parallel_enhanced_feature_generation(train_df[SMILES_COL].tolist(), num_cores)
test_features_list = parallel_enhanced_feature_generation(test_df[SMILES_COL].tolist(), num_cores)

# Convert to DataFrames
train_enhanced_features = pd.DataFrame(train_features_list).fillna(0)
test_enhanced_features = pd.DataFrame(test_features_list).fillna(0)

print(f"âœ… Generated {train_enhanced_features.shape[1]} enhanced features!")

# Ensure same columns in both datasets
common_features = list(set(train_enhanced_features.columns) & set(test_enhanced_features.columns))
train_enhanced_features = train_enhanced_features[common_features]
test_enhanced_features = test_enhanced_features[common_features]

# Remove constant features
constant_features = []
for col in train_enhanced_features.columns:
    if train_enhanced_features[col].nunique() <= 1:
        constant_features.append(col)

train_enhanced_features = train_enhanced_features.drop(columns=constant_features)
test_enhanced_features = test_enhanced_features.drop(columns=constant_features)

print(f"ğŸ—‘ï¸  Removed {len(constant_features)} constant features")

# Combine with existing features and target
train_ag = pd.concat([train_df.drop(columns=[SMILES_COL, ID_COL]), train_enhanced_features], axis=1)
test_ag = pd.concat([test_df.drop(columns=[SMILES_COL, ID_COL]), test_enhanced_features], axis=1)

print(f"âœ… Final dataset shapes:")
print(f"Train: {train_ag.shape}")
print(f"Test: {test_ag.shape}")
print(f"ğŸ¯ Enhanced feature engineering completed with {train_ag.shape[1]} total features!")
```

```python
predictor = TabularPredictor(
    label=TARGET,
    problem_type='regression',
    eval_metric='mae'
).fit(
    train_data=train_ag,
    time_limit=3600*9,
    presets='best_quality' 
)
```

```python
predictor.leaderboard(silent=True).style.background_gradient(subset=['score_val'], cmap='RdYlGn') 
```

```python
print("\nGenerating test predictions...")
test_preds = predictor.predict(test_ag) 
```
